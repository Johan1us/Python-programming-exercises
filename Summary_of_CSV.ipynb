```
import pandas as pd
```

```
name_csv = 'obj_kenm' 
```

```
df = pd.read_csv((name_csv + '.csv'), sep=";",  encoding='latin', error_bad_lines=False)
```

```
df.columns
```

```
df
```

```
def summary():
  print ('Summary of......:', name_csv + '.csv')
  print ('Columns.........:', len(df.columns))
  print ('Rows............:', len(df.index))
  print (2*('\n'))

  for col in df.columns:
    objcounts = df[(col)].value_counts()
    count_unique = len(objcounts[objcounts != 0])
    print ('  **', (col) ,'**''\n') # print the header of the column
    print ('Rows............:', len(df[(col)].index))
    print ('Filled..........:', df[(col)].count())
    print ('Empty...........:', df[(col)].isna().sum())
    print ('Unique values...:', count_unique)
 
    if df[(col)].count() > 1: # mistake: 'if objcounts.iloc[0] is > 1:'
      high, high_freq = objcounts.index[0], objcounts.iloc[0]
      low, low_freq = objcounts.index[-1], objcounts.iloc[-1] # lowest: trick = -1
      print ('Most freq.......:',high,'  (',high_freq,'x)')
      print ('Least freq......:',low,'  (',low_freq,'x)')
      print ('Datatype........:', df[(col)].dtype) # print the datatype of the column + a new line
      print ('Overview........:''\n')
      head = df[(col)][:5]
      for a in head:
        print ('    ',a)
      print ('     ...')
      tail = df[(col)][-5:]
      for b in tail:
        print ('    ',b)
      print (1*('\n'))
        
    else:
      print ('\n''** This is an empty column **''\n')
  
    if df[(col)].dtype == 'int64':
      objcounts = df[(col)].value_counts()
      count_unique = len(objcounts[objcounts != 0])
      high, high_freq = objcounts.index[0], objcounts.iloc[0]
    
      print ('Min.............:', df[(col)].min())
      print ('Mean............:', df[(col)].mean())
      print ('Max.............:', df[(col)].max())
      print ('Datatype........:', df[(col)].dtype) # print the datatype of the column + a new line
      print ('Overview........:''\n')
      head = df[(col)][:5]
      for a in head:
        print ('    ',a)
      print ('     ...')
      tail = df[(col)][-5:]
      for b in tail:
        print ('    ',b)
      print (1*('\n'))
  
  print (1*('\n'))
```

```
summary()
```
